```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score as r2
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('data.csv',sep=',')
print(df.head())
```

```python

#d1=df["id"]["diagnosis"]["radius_mean"]
#d1=df[:30]
dff=df.drop(["Unnamed: 32"],axis=1)#delete the empty column
#print (df.head())
#print("DFF")
#print (dff.head())
x=dff.drop(['diagnosis','id'],axis=1)
#x=dff.drop(['id'],axis=1)
print(x.head())
y = (dff["diagnosis"] == 'M').astype(np.int)
print(y)
y=y.reshape(-1,1)
#print(y)
#print (d1.head())
```

```python
from sklearn.metrics import accuracy_score as ac
from sklearn.cross_validation import train_test_split

Xt,Xl,yt,yl=train_test_split(x, y, test_size=0.2)

log_reg = LogisticRegression()
log_reg.fit(Xt, yt)
y_proba_train = log_reg.predict_proba(Xt)
y_pred_train=np.argmax(y_proba_train,axis=1)

Accuracy_train=ac(yt,y_pred_train)
print('Y train -> ',y_proba_train)
print('Accuracy Training ->',Accuracy_train)
```
